# Text Preprocessing


```python
import pickle
import pandas as pd
from nltk.stem.porter import PorterStemmer
```

```python
# Load Phase 3 dataframe
new_df = pickle.load(open("../artifacts/new_df_phase3.pkl", "rb"))

print("Loaded dataframe shape:", new_df.shape)
```

```python
new_df.head()
```

```python
# Create stemmer object
ps = PorterStemmer()
```

```python
# Stemming function
def stem(text):
    stemmed_words = []

    for word in text.split():
        stemmed_words.append(ps.stem(word))

    return " ".join(stemmed_words)
```

```python
# Apply stemming
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
# Check result
display(new_df.head())
```

```python
# Save final processed dataframe
pickle.dump(new_df, open("../artifacts/new_df_final.pkl", "wb"))

print("Final processed dataframe saved successfully!")
```
