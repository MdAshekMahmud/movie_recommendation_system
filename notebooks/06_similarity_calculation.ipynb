```python
import pickle
import pandas as pd

from sklearn.metrics.pairwise import cosine_similarity
```

## Load Final DataFrame and Vectors


```python
# Load final processed dataframe
new_df = pickle.load(open("../artifacts/new_df_final.pkl", "rb"))

# Load vectors from Phase 5
vectors = pickle.load(open("../artifacts/vectors.pkl", "rb"))

print("DataFrame shape:", new_df.shape)
print("Vectors shape:", vectors.shape)

display(new_df.head())
```

## Calculate Cosine Similarity


```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity[0]
```

```python
new_df.iloc[0]["title"]
```

```python
similarity[0][0]
```

## Sort Similar Movies


```python
movie_index = 0
similarity_score = list(enumerate(similarity[movie_index]))

similarity_score[:5]
```

```python
sorted_score = sorted(similarity_score, reverse=True, key=lambda x: x[1])
sorted_score[:10]
```

```python
# First result is the movie itself.So, we skip the first result and take the next
top_5_similar = sorted_score[1:6]

top_5_similar
```

## Display Top 5 Similar Movies


```python
movie_index = 0

print("Selected movie:", new_df.iloc[movie_index]["title"])
print("\nTop 5 similar movies:")

for i in sorted_score[1:6]:
    similar_movie_index = i[0]
    similarity_score = i[1]

    print(new_df.iloc[similar_movie_index]["title"], "->", similarity_score)
```

```python
# Save the Similarity Matrix
pickle.dump(similarity, open("../artifacts/similarity.pkl", "wb"))

print("Similarity matrix saved successfully!")
```

```python
# Save the final movie dataframe
pickle.dump(new_df, open("../artifacts/movies.pkl", "wb"))

print("Movies dataframe saved successfully!")
```
