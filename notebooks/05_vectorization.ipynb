# Vectorization


```python
import pickle
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
```

## Load the Final Processed DataFrame


```python
new_df = pickle.load(open("../artifacts/new_df_final.pkl", "rb"))

print("Loaded dataframe shape:", new_df.shape)

new_df.head()
```

```python
new_df.iloc[0]["title"]
```

```python
new_df.iloc[0]["tags"]
```

`max_features=5000`

This means we keep only the top 5000 most important/frequent words from the tags column.


```python
# Create the CountVectorizer Object
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
# Convert Tags into Vectors
vectors = cv.fit_transform(new_df["tags"]).toarray()
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
cv.get_feature_names_out()[:50]
```

```python
"avatar" in cv.get_feature_names_out()
```

```python
# Save the Vectorizer and Vectors
pickle.dump(cv, open("../artifacts/count_vectorizer.pkl", "wb"))
pickle.dump(vectors, open("../artifacts/vectors.pkl", "wb"))

print("Vectorizer and vectors saved successfully!")
```

## Check a Movie’s Non-Zero Words


```python
movie_index = 0

feature_names = cv.get_feature_names_out()
movie_vector = vectors[movie_index]

non_zero_indices = movie_vector.nonzero()[0]

print("Movie title:", new_df.iloc[movie_index]["title"])

print("\nWords present in this movie vector:")
for index in non_zero_indices:
    print(feature_names[index], ":", movie_vector[index])
```
