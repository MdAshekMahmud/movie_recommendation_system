# Model Saving and Final Artifact Management


```python
import os
import pickle
import pandas as pd
```

```python
# Load Existing Final Files
artifacts_dir = "../artifacts"

new_df = pickle.load(open(os.path.join(artifacts_dir, "new_df_final.pkl"), "rb"))
cv = pickle.load(open(os.path.join(artifacts_dir, "count_vectorizer.pkl"), "rb"))
vectors = pickle.load(open(os.path.join(artifacts_dir, "vectors.pkl"), "rb"))
similarity = pickle.load(open(os.path.join(artifacts_dir, "similarity.pkl"), "rb"))
movies = pickle.load(open(os.path.join(artifacts_dir, "movies.pkl"), "rb"))
```

```python
print("new_df shape:", new_df.shape)
print("vectors shape:", vectors.shape)
print("similarity shape:", similarity.shape)
print("movies shape:", movies.shape)
```

```python
# Check Artifact Alignment
assert (
    movies.shape[0] == similarity.shape[0]
), "Movies and similarity row count mismatch"
assert movies.shape[0] == similarity.shape[1], "Similarity matrix column count mismatch"

print("Artifacts are aligned correctly.")
```

```python
# Save Final App Artifacts Again
movies = movies.reset_index(drop=True)

pickle.dump(movies, open(os.path.join(artifacts_dir, "movies.pkl"), "wb"))
pickle.dump(similarity, open(os.path.join(artifacts_dir, "similarity.pkl"), "wb"))

print("Final app artifacts saved successfully.")
```

```python
# Create Artifact Test Function
def test_artifacts():
    movies = pickle.load(open("../artifacts/movies.pkl", "rb"))
    similarity = pickle.load(open("../artifacts/similarity.pkl", "rb"))

    movies = movies.reset_index(drop=True)

    print("Movies shape:", movies.shape)
    print("Similarity shape:", similarity.shape)

    assert "title" in movies.columns, "title column missing"
    assert movies.shape[0] == similarity.shape[0], "Row mismatch"
    assert movies.shape[0] == similarity.shape[1], "Column mismatch"

    print("Artifact test passed successfully.")


test_artifacts()
```
