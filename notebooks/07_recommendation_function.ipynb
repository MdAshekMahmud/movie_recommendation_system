# Recommendation Function


```python
import pickle
import pandas as pd
```

```python
# Load Saved Artifacts
movies = pickle.load(open("../artifacts/movies.pkl", "rb"))
similarity = pickle.load(open("../artifacts/similarity.pkl", "rb"))
```

```python
print("Movies shape:", movies.shape)
print("Similarity matrix shape:", similarity.shape)

movies.head()
```

```python
# Reset Index Safely
movies = movies.reset_index(drop=True)
```

```python
# Find a movie inde
movie_name = "Avatar"

movie_index = movies[movies["title"] == movie_name].index[0]
movie_index
```

```python
# Get Similarity Score for that Movie
similarity_scores = list(enumerate(similarity[movie_index]))
similarity_scores[:5]
```

```python
# Basic Recommendation Function
def recommend(movie_name, top_n=5):
    movie_name = movie_name.strip().lower()

    matches = movies[movies["title"].str.lower() == movie_name]

    if matches.empty:
        return f"Movie '{movie_name}' not found in the dataset."

    movie_index = matches.index[0]

    similarity_scores = list(enumerate(similarity[movie_index]))

    sorted_scores = sorted(similarity_scores, reverse=True, key=lambda x: x[1])

    recommendations = []

    for item in sorted_scores[1 : top_n + 1]:
        recommended_movie_index = item[0]
        recommended_title = movies.iloc[recommended_movie_index]["title"]
        similarity_score = item[1]

        recommendations.append(
            {"movie": recommended_title, "similarity_score": round(similarity_score, 4)}
        )

    return pd.DataFrame(recommendations)
```

```python
recommend("Avatar")
```

```python
recommend("The dark knight", top_n=6)
```

```python
recommend("Random Fake Movie", top_n=5)
```

## Better Partial Search Function


```python
def search_movie(query):
    query = query.strip().lower()

    results = movies[movies["title"].str.lower().str.contains(query, na=False)]

    return results[["title"]].head(20)
```

```python
search_movie("dark knight")
```

```python
search_movie("spider")
```
