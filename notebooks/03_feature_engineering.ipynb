# Feature Engineering


```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv("../data/tmdb_5000_movies.csv")
credits = pd.read_csv("../data/tmdb_5000_credits.csv")

# Merge datasets
movies = movies.merge(credits, on="title")

# Select useful columns
movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]

# Remove missing values
movies.dropna(inplace=True)

# Remove duplicates if any
movies.drop_duplicates(inplace=True)

movies.head()
```

```python
movies.iloc[0]["genres"]
```

```python
# Create Function to Extract Names


def convert(obj):
    names = []

    for item in ast.literal_eval(obj):
        names.append(item["name"])

    return names
```

```python
movies["genres"] = movies["genres"].apply(convert)
movies["keyworkds"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
# Extract Top 3 Cast Members


def convert_cast(obj):
    names = []
    counter = 0

    for item in ast.literal_eval(obj):
        if counter != 3:
            names.append(item["name"])
            counter += 1

        else:
            break

    return names
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
```

```python
movies.iloc[0]["cast"]
```

```python
# Extract Director from Crew


def fetch_director(obj):
    directors = []

    for item in ast.literal_eval(obj):
        if item["job"] == "Director":
            directors.append(item["name"])
            break
    return directors
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.iloc[0]["crew"]
```

## Convert Overview into Word List

`overvies` is a sentence.

we need to convert it into a list of works.


```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.iloc[0]["overview"]
```

## Remove Spaces Inside Names

Suppose we keep:

Sam Worthington

CountVectorizer will treat this as two separate words:

Sam

Worthington


```python
def remove_space(word_list):
    new_list = []

    for word in word_list:
        new_list.append(word.replace(" ", ""))

    return new_list
```

```python
movies["genres"] = movies["genres"].apply(remove_space)
movies["keywords"] = movies["keywords"].apply(remove_space)
movies["cast"] = movies["cast"].apply(remove_space)
movies["crew"] = movies["crew"].apply(remove_space)
```

```python
movies.head()
```

## Create the tags Column

overview + genres + keywords + cast + crew


```python
movies["tags"] = (
    movies["overview"]
    + movies["genres"]
    + movies["keywords"]
    + movies["cast"]
    + movies["crew"]
)
```

```python
movies.iloc[0]["tags"]
```

## Create Final DataFrame


```python
new_df = movies[["movie_id", "title", "tags"]]
```

```python
new_df.head()
```

```python
# Convert Tags List into String
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df.iloc[0]["tags"]
```

```python
# Convert Tags to Lowercase
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
new_df.iloc[0]["tags"]
```
