# Dataset Understanding

This notebook performs a brief exploratory data analysis on the
TMDB 5000 Movie Dataset. It loads the movies and credits files,
inspects their shapes, columns, missing values and a few example
rows.


```python
import numpy as np
import pandas as pd
```

```python
# Load the TMDB datasets
movies = pd.read_csv("../data/tmdb_5000_movies.csv")
credits = pd.read_csv("../data/tmdb_5000_credits.csv")
```

```python
print("Movies dataset shape:", movies.shape)
print("Credits dataset shape:", credits.shape)
```

```python
# Display the first five rows of each dataset
display(movies.head(3))
display(credits.head(3))
```

```python
print("Movies columns:")
print(movies.columns)

print("Credits columns:")
print(credits.columns)
```

```python
print("Movies info:")
movies.info()
```

```python
print("Credits info:")
credits.info()
```

```python
print("Missing values in movies dataset:")
print(movies.isnull().sum())
```

```python
print("Missing values in credits dataset:")
print(credits.isnull().sum())
```

```python
print("Duplicate rows in movies:", movies.duplicated().sum())
print("Duplicate rows in credits:", credits.duplicated().sum())
```
